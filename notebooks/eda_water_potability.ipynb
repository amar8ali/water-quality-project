# Exploratory Data Analysis (EDA) for Water Potability Dataset

This notebook explores the water potability dataset, including data loading, summary statistics, missing value analysis, and feature visualization.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the water potability dataset
# Update the path if needed
file_path = "../data/water_potability.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Show summary statistics
summary = df.describe()
summary
```

```python
# Visualize missing values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values in Water Potability Dataset')
plt.show()
```

```python
# Plot distributions of key features
features = ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()
for i, feature in enumerate(features):
    sns.histplot(df[feature].dropna(), ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

## Additional Visualizations

```python
# Correlation heatmap
plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Boxplots for feature distributions by potability
plt.figure(figsize=(15,10))
for i, feature in enumerate(features):
    plt.subplot(3, 3, i+1)
    sns.boxplot(data=df, x='Potability', y=feature)
    plt.title(f'{feature} by Potability')
plt.tight_layout()
plt.show()
```

```python
# Countplot of potability classes
sns.countplot(data=df, x='Potability')
plt.title('Potability Class Distribution')
plt.show()
```
